2.  What are the key assumptions of Simple Linear Regression?

-> The key assumptions of Simple Linear Regression (SLR) ensure that the model produces valid, unbiased, and reliable results. If these assumptions are violated, the regression estimates (slope, intercept) may not be accurate.

Here are the five main assumptions :    

 1. **Linearity**

There should be a linear relationship between the independent variable (X) and the dependent variable (Y).

This means that changes in X are associated with proportional changes in Y.

You can check this using a scatter plot â€” the points should roughly form a straight-line pattern.

 Example: As hours studied increase, exam score increases in a roughly straight-line manner.

 2. **Independence of Errors**

The residuals (errors) â€” differences between observed and predicted Y values â€” should be independent of each other.

No correlation should exist among residuals.

This assumption is often tested using the Durbin-Watson test.

 Violation Example: Time series data where one observation depends on the previous one (autocorrelation).

 3. **Homoscedasticity**

The residuals should have constant variance across all levels of X.

In other words, the spread of residuals should be roughly the same for all predicted values.

You can check this with a residuals vs. fitted values plot â€” the spread should be even (no funnel shape).

Violation Example: When variance increases with X (heteroscedasticity), predictions become less reliable.

 4. **Normality of Errors**

The residuals (errors) should be normally distributed.

This is important for hypothesis testing (e.g., testing significance of coefficients).

You can check this with a histogram or Q-Q plot of residuals.

 5. **No (or minimal) Measurement Error in X**

The independent variable (X) should be measured accurately and without error.

In SLR, itâ€™s assumed that all error lies in Y (the dependent variable).

Measurement error in X can bias the slope estimate.

9.  Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.



```python
# Import necessary libraries
import numpy as np
from sklearn.linear_model import LinearRegression

# Sample data
# Independent variable (X): Study hours
# Dependent variable (y): Exam scores
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # reshape for sklearn (2D array)
y = np.array([40, 50, 60, 65, 80])

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Get slope (Î²1) and intercept (Î²0)
slope = model.coef_[0]
intercept = model.intercept_

# Print the results
print("Slope (Î²1):", slope)
print("Intercept (Î²0):", intercept)

# Optional: Predict values
y_pred = model.predict(X)
print("Predicted values:", y_pred)
```
